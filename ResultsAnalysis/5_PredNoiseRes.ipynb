```python
df_price_all = pd.read_csv("../data/df_prices_c.csv", parse_dates = ["date"])
df_price_all["hour"] = df_price_all["date"].dt.hour
df_price_all["month"] = df_price_all["date"].dt.month
df_price_all["year"] = df_price_all["date"].dt.year
df_price = df_price_all[df_price_all["year"] == 2019]

df_price.info()
```

```python
def calc_MASE(df_price_og, noise_scale, runs = 30):
    df_price = df_price_og.copy()
    l_mase = []
    naive_mae =np.mean(np.abs(df_price["price_im"].shift(1) - df_price["price_im"]))

    for _ in range(runs):
        df_price["price_noise"] = df_price["price_im"] + np.random.normal(0, noise_scale, len(df_price))
        error = np.mean(np.abs(df_price["price_noise"] - df_price["price_im"]))
        l_mase.append(error / naive_mae)
    return np.mean(l_mase)

def calc_MAE(df_price_og, noise_scale, runs = 30):
    df_price = df_price_og.copy()
    l_mape = []

    for _ in range(runs):
        df_price["price_noise"] = df_price["price_im"] + np.random.normal(0, noise_scale, len(df_price))
        error = np.mean(np.abs(df_price["price_noise"] - df_price["price_im"]))
        l_mape.append(error)
    return np.mean(l_mape)

def calc_R2(df_price_og, noise_scale, runs = 30):
    df_price = df_price_og.copy()
    TSS = np.sum((df_price["price_im"] - df_price["price_im"].mean())**2)
    l_r2 = []
    for _ in range(runs):
        df_price["price_noise"] = df_price["price_im"] + np.random.normal(0, noise_scale, len(df_price))
        RSS = np.sum((df_price["price_noise"] - df_price["price_im"])**2)
        l_r2.append(1 - RSS/TSS)
    return np.mean(l_r2)
```
