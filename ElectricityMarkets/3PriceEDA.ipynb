```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df_price= pd.read_csv('data/df_price_2014_2019.csv', parse_dates=["date"])
df_price["year"] = df_price["date"].dt.year
df_price["month"] = df_price["date"].dt.month
df_price["hour"] = df_price["date"].dt.hour
df_price.info()
```

```python
years = df_price["year"].unique()
n_years = len(years)
fig1 = plt.figure(figsize = (20,20))
axs = fig1.subplots(n_years, 1, sharex=True)

for i, year in enumerate(years):
    df_year = df_price[df_price["year"] == year]
    print(f"{year=}, {len(df_year)=}")
    axs[i].hist(df_year["price_im"], bins = 100)
    axs[i].set_title(f"{year=}")
    
```

```python
fig1 = plt.figure(figsize = (18,18))
ax1 = fig1.add_subplot(4,1,1)
ax2 = fig1.add_subplot(4,1,2)
ax3 = fig1.add_subplot(4,1,3)
ax4 = fig1.add_subplot(4,1,4)

sns.boxplot(x="year", y="price_im", data=df_price, ax=ax1, hue="month", showfliers=False)
sns.move_legend(ax1, "lower center", ncol = 12)

sns.boxplot(x="month", y="price_im", data=df_price, ax=ax2, hue="year", showfliers=False)
sns.move_legend(ax2, "lower center", ncol = 12)

sns.boxplot(x="year", y="price_im", data=df_price, ax=ax3, hue="hour", showfliers=False)
sns.move_legend(ax3, "lower center", ncol = 12)

sns.boxplot(x="hour", y="price_im", data=df_price, ax=ax4, hue="year", showfliers=False)
sns.move_legend(ax4, "lower center", ncol = 12)
```

```python
fig1 = plt.figure(figsize = (18,6))
ax1 = fig1.add_subplot(1,1,1)
sns.boxplot(x="year", y="price_im", data=df_price, ax= ax1, hue="hour", showfliers=False)
sns.move_legend(ax1, "lower center", ncol = 12)
ax1.set_ylim([-0.15, 0.3])
```
